```python
movie_names = movies[['movieId', 'title']]

combined_movies_data = pd.merge(ratings, movie_names, on='movieId')

combined_movies_data = combined_movies_data[['userId','title']]

combined_movies_data.head()
```

```python
onehot = combined_movies_data.pivot_table(index='userId', columns='title', aggfunc=len, fill_value=0)
onehot = onehot.fillna(0)
onehot[onehot>1]=1
onehot.head()
```

```python
onehot.shape
```

```python
from mlxtend.frequent_patterns import association_rules, apriori

frequent_itemsets = apriori(onehot, min_support = 0.15, max_len = 2, use_colnames=True)

rules = association_rules(frequent_itemsets)
rules
```

```python
rules.sort_values(by = 'lift' ,ascending = False).head(10)
```

```python
rules[rules.antecedents.apply(str).str.contains("(Beauty and the Beast (1991))")].sort_values('lift', ascending=False)
```
